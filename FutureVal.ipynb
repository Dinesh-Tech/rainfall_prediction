# Imports

```python
import pandas as pd
import numpy as np
```

# Read Input 

The input file is present as inp.csv

```python
df = pd.read_csv('inp.csv')
df = df.dropna(how='all')
df.index = df['YEAR']
df = df.drop('YEAR', axis=1)
df
```

# Fill empty rows for years that are absent

```python
min_year = int(min(df.index))
max_year = int(max(df.index))
print(min_year, max_year)
df_all_years = df.reindex(list(range(min_year, max_year+1+2)))
df
```

# Compute future Values

```python
futurevals = df_all_years.copy()
futurevals.index = list(range(min_year-2, max_year+1))
futurevals = futurevals.loc[min_year:max_year-2,:]
futurevals
```

# Add prefix to values in futurevals

```python
for month in futurevals.columns:
    for year in futurevals.index:
        futurevals.loc[year, month] = f"{year+2}-{month}-{futurevals.loc[year, month]}"
```
